This is tutoriaal describes to necessary steps to set up OpenMP taks loading on Google Colab. These steps will need to be repeated for each notebook and each time a notebook is launched.  

### GCC 

By default the Colab VM comes with GCC 7.5.0

```python
!gcc --version
```

1. For target offloading, we will need to install GCC 8 or later. Shouldn't take more than 10 seconds via `apt` 

```python
!apt install gcc-8 g++-8 gfortran-8
```

2. We then need to install the NVIDIA offloading package. 

```python
!apt install gcc-8-offload-nvptx
```

3. Now make GCC 8 the default.

```python
!update-alternatives --install /usr/bin/gcc gcc /usr/bin/gcc-8 80 --slave /usr/bin/g++ g++ /usr/bin/g++-8
```

4. Verify the default is set to version 8. 

```python
!gcc --version 
```


### NVIDIA driver 

By default Colab uses NVIDIA 11 (last check 07/14/2022). 

```python
!nvcc --version
```

NVIDIA 11 has deprecated CUDA architecture `SM_30`. Unfortunately, this breaks builds for GCC-OpenMP offloading. GCC8 does not provide a flag for selecting a particular GPU architecture. 

5. The simplest workaround is to rollback to CUDA 10

```python
!rm -rf /usr/local/cuda
!ln -s /usr/local/cuda-10.0 /usr/local/cuda
```

```python
!nvcc --version 
```
